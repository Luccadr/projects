# Construindo o ML

```python
df_quant = df.groupby('fullVisitorId',as_index=False)[quant].sum()
```

```python
df_quant.head()
```

```python
y = df_quant.transactionRevenue.copy()
```

```python
X = df_quant.drop('transactionRevenue',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

## Regressão Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg_predict = reg.predict(X_test)
```

```python
reg_predict
```

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['erro'] = reg_predict - y_test
```

```python
resultados.head()
```

```python
# Checando se alguma predição é negativ ( o que nã faria sentido)
resultados[resultados.predict < 0]
```

```python
import numpy as np
```
