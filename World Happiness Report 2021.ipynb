# World Happiness Report 2021

## Factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

### Key Features in 'world happiness report':

#### *Life Ladder: Imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you.

#### *Log GDP per capita: At its most basic interpretation, per capita GDP shows how much economic production value can be attributed to each individual citizen. Alternatively, this translates to a measure of national wealth since GDP market value per person also readily serves as a prosperity measure.

#### *Social support: Social support is defined in terms of social network characteristics such as assistance from family, friends, neighbours and other community members.

#### *Healthy life expectancy at birth: Healthy life expectancy is the average life in good health - that is to say without irreversible limitation of activity in daily life or incapacities - of a fictitious generation subject to the conditions of mortality and morbidity prevailing that year

#### *Freedom to make life choices: Freedom of choice describes an individual's opportunity and autonomy to perform an action selected from at least two available options, unconstrained by external parties.

#### *Generosity: Generosity is the virtue of being liberal in giving, often as gifts.

#### *Perceptions of corruption: Corruption is a form of dishonesty or criminal offense undertaken by a person or organization entrusted with a position of authority, to acquire illicit benefit

```python
median_GDP = sts.median(data['Log GDP per capita'])
median_GDP
```

```python
#Using the method to fillna, using mean as information
median_social = sts.median(data['Social support'])
median_healthy = sts.median(data['Healthy life expectancy at birth'])
median_freedom = sts.median(data['Freedom to make life choices'])
median_generosity = sts.median(data['Generosity'])
median_corruption = sts.median(data['Perceptions of corruption'])
median_positive = sts.median(data['Positive affect'])
median_negative = sts.median(data['Negative affect'])
```

```python
data['Social support'].fillna(median_social, inplace = True)
data['Healthy life expectancy at birth'].fillna(median_healthy, inplace = True)
data['Freedom to make life choices'].fillna(median_freedom, inplace = True)
data['Generosity'].fillna(median_generosity, inplace = True)
data['Perceptions of corruption'].fillna(median_corruption, inplace = True)
data['Positive affect'].fillna(median_positive, inplace = True)
data['Negative affect'].fillna(median_negative, inplace = True)
```

### Progression and Evolution of Countries by Decade

```python
# Creating a columns for decades information 

data_compare_decade10 = data_compare_decade.loc[data_compare_decade['Decade'] == 10].set_index("Country name")
data_compare_decade11 = data_compare_decade.loc[data_compare_decade['Decade'] == 11].set_index("Country name")
```

```python
data_compare_decade10
```

```python
data_compare_decade11
```

```python
data_compare_decade10["Life Ladder 10"] = data_compare_decade10["Life Ladder"]
data_compare_decade11["Life Ladder 11"] = data_compare_decade11["Life Ladder"]
```

```python
data_compare_decade10
```

```python
# Mergeing the data reated with decade 11 with 10

data_evolution = data_compare_decade11.merge(data_compare_decade10, how = 'left',left_index = True,
                                            right_on = "Country name")
data_evolution
```

```python
# Checking the countries that don't have information for some decade

data_evolution = data_evolution.loc[:,("Country name","Life Ladder 10","Life Ladder 11")]
data_evolution.isnull().sum()
```

```python
countries_removed = data_evolution[data_evolution["Life Ladder 10"].isnull()]
countries_removed = pd.DataFrame(countries_removed.iloc[:,0])
countries_removed
```

```python
data_evolution.dropna(inplace = True)
data_evolution.drop(columns = ["Country name"], inplace = True)
```

```python
data_evolution.rename(columns={"Life Ladder 11" : "Evolution" ,"Life Ladder 10" : 10}, inplace = True)
```

```python
pct_evolution_data = data_evolution.pct_change(axis='columns')
pct_evolution_data = pct_evolution_data.drop(columns = 10)
```

```python
pct_evolution_data
```

```python
pct_evolution_data.rename(columns = {"11" : "Evolution"}, inplace = True)
#data_evolution.rename(columns={"Life Ladder 11" : 11 ,"Life Ladder 10" : 10}, inplace = True)
```
